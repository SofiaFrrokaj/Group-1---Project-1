```python
#import necessary libraries
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
#read the csv file and store it in a dataframe
crime_data = pd.read_csv(Path('crime_data.csv'))
```

```python
#display the first few rows of the data

crime_data.head()
```

```python
#display the column names
crime_data.keys()
```

```python
# remove rows with missing values
crime_data=crime_data.dropna()
```

```python
#display the last few rows of the data
crime_data.tail()
```

```python
#rename the columns names for better understanding
crime_data=crime_data.rename(columns={'CMPLNT_NUM':"Complaint ID",
                                       'CMPLNT_FR_DT': "Crime Date",
                                       'CMPLNT_FR_TM':"Complaint Time",
                                         'CMPLNT_TO_DT':"Ending Date",
                                        'CMPLNT_TO_TM':"Ending Time", 
                                        'RPT_DT':"Reported Date",
                                        'KY_CD':"Offense Code",
                                        'OFNS_DESC':"Offense Description",
                                        'PD_CD':"Classification Code",
                                        'PD_DESC':"Classification Description",
                                        'CRM_ATPT_CPTD_CD':"Crime Status",
                                        'LAW_CAT_CD':"Offense Level", 
                                        'JURIS_DESC':"Jurisdiction",
                                        'BORO_NM':"Borough",
                                        'ADDR_PCT_CD':"Precinct",
                                        'LOC_OF_OCCUR_DESC':"Location Description", 
                                        'PREM_TYP_DESC':"Premise Description",
                                        'PARKS_NM':"Park Name",
                                        'HADEVELOPT':"Name of NYCHA Housing", 
                                        'X_COORD_CD':"X Coordinate", 
                                        'Y_COORD_CD':"Y Coordinate",
                                        'Latitude':"Latitude",
                                         'Longitude':"Longitude",
                                        })
      
```

```python
#display the first few rows of the data 
crime_data.head()
```

```python
#display the column names
crime_data.columns
```

```python
#What percentage of crimes result in arrests?
# Drop rows with missing values
df_crime_data=crime_data.dropna()

# Count total crimes
total_crimes = len(df_crime_data)

# Count crimes that resulted in arrests (assuming "Arrest" is the relevant status)
arrested_crimes = df_crime_data[df_crime_data["Crime Status"] == "COMPLETED"]

# Calculate percentage
arrest_rate = (len(arrested_crimes) / total_crimes) * 100
# Print the result
print(f"Percentage of crimes resulting in arrests: {arrest_rate:.2f}%")
```

```python
#Which offenses are the most common?
# Count the occurrences of each offense
common_offenses = df_crime_data["Offense Description"].value_counts().head(10)
display (common_offenses)
# Plot the top 10 most common offenses
common_offenses.plot(kind="bar", title="Top 10 Most Common Offenses")


#how to make the plot more readable

# Plot the top 10 most common offenses
common_offenses.plot(kind="bar", title="Top 10 Most Common Offenses")
plt.xticks
plt.ylabel("Number of Crimes")
plt.xlabel("Offense Description")
plt.show()
```

```python
#Which offenses are the least common?
# Count the occurrences of each offense
rare_offenses = df_crime_data["Offense Description"].value_counts().tail(10)
display (rare_offenses)
# Plot the top 10 least common offenses
rare_offenses.plot(kind="bar", title="Top 10 Least Common Offenses")
plt.xticks
plt.ylabel("Number of Crimes")
plt.xlabel("Offense Description")
plt.show()

```

```python
#What we can learn from the above analysis is that:
#The most common offenses are related to drugs and weapons, while the least common offenses are related to forgery and offenses involving fraudulent activities. 
# This information could be useful for law enforcement agencies to allocate resources more effectively.
```

```python
#Which locations are the most common for crimes to occur?
# Count the occurrences of each location
common_locations = df_crime_data["Premise Description"].value_counts().head()
display (common_locations)
# Plot the top locations
common_locations.plot(kind="bar", title="Top Locations for Crimes")
```

```python
#How does the clearance rate (solved cases) vary by crime type?
#Count the clearance rate by offense type

df_clearance_rate = df_crime_data.groupby("Offense Description")["Crime Status"].value_counts(normalize=True).loc[:, "COMPLETED"].sort_values(ascending=False)
display(df_clearance_rate)
# Plot the clearance rate by offense type
df_clearance_rate.plot(kind="barh", title="Clearance Rate by Crime Type")
plt.xlabel("Clearance Rate")
plt.ylabel("Offense Description")
plt.show()
```

```python
#How does the clearance rate vary by borough?
# Count the clearance rate by borough
df_clearance_rate = df_crime_data.groupby("Borough")["Crime Status"].value_counts(normalize=True).loc[:, "COMPLETED"].sort_values(ascending=False)
display(df_clearance_rate)
# Plot the clearance rate by borough
df_clearance_rate.plot(kind="bar", title="Clearance Rate by Borough")
plt.xlabel("Borough")
plt.ylabel("Clearance Rate")
plt.show()
```

```python
#How does the clearance rate vary by offense level?
# Count the clearance rate by offense level
df_clearance_rate = df_crime_data.groupby("Offense Level")["Crime Status"].value_counts(normalize=True).loc[:, "COMPLETED"].sort_values(ascending=False)
display(df_clearance_rate)
# Plot the clearance rate by offense level
df_clearance_rate.plot(kind="bar", title="Clearance Rate by Offense Level")
plt.xlabel("Offense Level")
plt.ylabel("Clearance Rate")
plt.show()
```

```python
#How does the clearance rate vary by premise description?
# Count the clearance rate by premise description
df_clearance_rate = df_crime_data.groupby("Premise Description")["Crime Status"].value_counts(normalize=True).loc[:, "COMPLETED"].sort_values(ascending=False)
display(df_clearance_rate)
# Plot the clearance rate by premise description
df_clearance_rate.plot(kind="bar", title="Clearance Rate by Premise Description")
plt.xlabel("Premise Description")
plt.ylabel("Clearance Rate")
plt.show()
```

```python
#How does the clearance rate vary by jurisdiction?
# Count the clearance rate by jurisdiction
df_clearance_rate = df_crime_data.groupby("Jurisdiction")["Crime Status"].value_counts(normalize=True).loc[:, "COMPLETED"].sort_values(ascending=False)
display(df_clearance_rate)
# Plot the clearance rate by jurisdiction
df_clearance_rate.plot(kind="bar", title="Clearance Rate by Jurisdiction")
plt.xlabel("Jurisdiction")
plt.ylabel("Clearance Rate")
plt.show()
```
